def SpeechToText(file_name):
  client = speech.SpeechClient.from_service_account_file('google_secret_key.json')
  file_name = "Data/audio/"+ f'{file_name}'
  recognizer = speechtotext.Recognizer()
  with speechtotext.AudioFile(file_name) as source:
      audio_text = recognizer.listen(source)

  with open(file_name, 'rb') as f:
    mp3_data = f.read()
  audio_file = speech.RecognitionAudio(content=mp3_data)

  config = speech.RecognitionConfig(
      encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
      sample_rate_hertz=44100,
      language_code="en-US",
      audio_channel_count=2,
      enable_separate_recognition_per_channel=True,
  )
  try:
      # using google speech to text API
      text = recognizer.recognize_google(audio_text)
      print('Converting audio transcripts into text ...')
      print('Extracted Text: '+text)
      file.write('Extracted Text: '+text+'\n')

      translator = Translator()

      translated_text = (translator.translate(text, dest='zh-tw')).text
      print('Chinese: '+ translated_text)
      file.write('Chinese: '+translated_text+'\n')
      
  except:
        print('Sorry.. run again...')

這段程式碼定義了一個函數 SpeechToText(file_name)，它會將音頻文件轉換為文字並將其翻譯為中文。

1.首先，它使用 Google Cloud Speech API 建立一個 client，並使用 service account 的 json 檔案來驗證使用者。
2.接著，它使用 speech_recognition 套件來聆聽音頻文件，並使用 recognize_google 方法將音頻轉換為文字。
3.然後，它使用 googletrans 套件將文字翻譯為中文。
4.最後，它將文字和翻譯後的文字寫入 Result.txt 檔案中。

如果執行途中出現錯誤，就會顯示 "Sorry.. run again..."。

```
def SpeechToText(file_name):
  client = speech.SpeechClient.from_service_account_file('google_secret_key.json')
  file_name = "Data/audio/"+ f'{file_name}'
  recognizer = speechtotext.Recognizer()
  with speechtotext.AudioFile(file_name) as source:
      audio_text = recognizer.listen(source)

  with open(file_name, 'rb') as f:
    mp3_data = f.read()
  audio_file = speech.RecognitionAudio(content=mp3_data)

  config = speech.RecognitionConfig(
      encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
      sample_rate_hertz=44100,
      language_code="en-US",
      audio_channel_count=2,
      enable_separate_recognition_per_channel=True,
  )
  try:
      # using google speech to text API
      text = recognizer.recognize_google(audio_text)
      print('Converting audio transcripts into text ...')
      print('Extracted Text: '+text)
      file.write('Extracted Text: '+text+'\n')

      translator = Translator()

      translated_text = (translator.translate(text, dest='zh-tw')).text
      print('Chinese: '+ translated_text)
      file.write('Chinese: '+translated_text+'\n')
      
  except:
        print('Sorry.. run again...')
```
